```python
def  plot_im(im):
    import matplotlib.pyplot as plt
    %matplotlib inline
    import numpy as np
    fig=plt.figure(figsize=(8,6))
    #fig.clf()
    ax=fig.gca()
    ax.imshow(np.array(im).astype(int))
    return 'Done'
 
```

## Explore more  images  from a file 

```python
import os
import glob
os.chdir('C:/Users/alexandru.daia/Desktop/TAIE')
```

```python
def read_memories(path='TAIE',filename="*.jpg"):
    import os
    from scipy import misc
    image_list=[]
    file_path=os.path.join(path,filename)
    for filename in  glob.glob(file_path):
        im=misc.imread(filename)
        image_list.append(im)
    return image_list
    
```

```python
image_list=read_memories()
```
