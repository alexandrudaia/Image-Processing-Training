```python
def hist(im,bins=256):
    """Display histogram of  flatten image"""
    import matplotlib.pyplot as plt
    import numpy as np
    fig=plt.figure(figsize=(8,6))
    fig.clf()
    ax=fig.gca()
    ax.hist(np.array(im).flatten(),bins=bins)
```

```python
hist(im)
```

```python
def  cdf(im,bins=256):
    import matplotlib.pyplot as plt
    import numpy as np
    y,x=np.histogram(np.array(im).flatten(),bins=bins)
    y=y.cumsum()
    fig=plt.figure(figsize=(8,6))
    fig.clf()
    ax=fig.gca()
    ax.plot(x[:256],y)
    
cdf(im)
```

```python
import numpy as np
def image_equalize(im):
    """Function  to equalize  the image"""
    from skimage  import exposure 
    out=exposure.equalize_adapthist(im)
    out *=255.0/np.amax(out)# normalize
    return out.astype(np.uint8)
```

```python
im_eq=image_equalize(im)
```

```python
hist(im_eq)
```

```python
cdf(im_eq)
```

```python
def plot(im1, im2):
    import matplotlib.pyplot as plt
    import numpy  as np
    fig,ax=plt.subplots(1,2,figsize=(12,12))
    ax[0].imshow(im1)
    ax[1].imshow(im2)
    return 'Done'
```

```python
plot(im,im_eq)
```

# Bluring / denoising with   gaussian   filter convolution  or median filter

## Median filters  tend to  be better  at  preserving edges then  gaussian convolution

```python
def gauss_filter(im,sigma=5):
    from scipy.ndimage.filters  import gaussian_filter as  gf
    import numpy as np
    im_a=np.array(im)
    return(gf(im_a,sigma=sigma))
```

```python
im_g=gauss_filter(im_eq)
hist(im_g)
```

```python
cdf(im_g)
```

```python
plot(im_eq,im_g)
```

```python
def med_filter(im,size=4):
    from scipy.ndimage.filters import median_filter as mf
    import numpy as np
    return mf(im,size=size)
```

```python
im_m=med_filter(im)
```

```python
plot(im_eq,im_m)
```
