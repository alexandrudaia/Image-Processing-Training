```python
def read_memories(path='TAIE',filename="*.jpg"):
    import os
    from scipy import misc
    image_list=[]
    file_path=os.path.join(path,filename)
    for filename in  glob.glob(file_path):
        im=misc.imread(filename)
        image_list.append(im)
    return image_list
    
```

## Image properties

```python
def hist(im,bins=256):
    """Display histogram of  flatten image"""
    import matplotlib.pyplot as plt
    import numpy as np
    fig=plt.figure(figsize=(8,6))
    fig.clf()
    ax=fig.gca()
    ax.hist(np.array(im).flatten(),bins=bins)
```

```python
hist(im)
```

```python
#it is bumpy meaning  there could  be   not enough saturation of   some levels
```

```python
def  cdf(im,bins=256):
    import matplotlib.pyplot as plt
    import numpy as np
    y,x=np.histogram(np.array(im).flatten(),bins=bins)
    y=y.cumsum()
    fig=plt.figure(figsize=(8,6))
    fig.clf()
    ax=fig.gca()
    ax.plot(x[:256],y)
    
cdf(im)
```

```python
import numpy as np
def image_equalize(im):
    """Function  to equalize  the image"""
    from skimage  import exposure 
    out=exposure.equalize_adapthist(im)
    out *=255.0/np.amax(out)# normalize
    return out.astype(np.uint8)
```

```python
grandPadre_EQ=image_equalize(grandPadre)
```

```python
hist(grandPadre_EQ)
```

```python
cdf(grandPadre_EQ)
```

```python
def plot(im1, im2):
    import matplotlib.pyplot as plt
    import numpy  as np
    fig,ax=plt.subplots(1,2,figsize=(12,12))
    ax[0].imshow(im1)
    ax[1].imshow(im2)
    return 'Done'
```

```python
plot(grandPadre,grandPadre_EQ)
```
